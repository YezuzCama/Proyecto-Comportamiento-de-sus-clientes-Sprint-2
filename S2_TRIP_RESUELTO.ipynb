# Paso 1: Crear Función clean_user

En la última tarea de la primera parte de este proyecto escribiste código para:

1. Eliminar todos los espacios iniciales y finales de los nombres, así como cualquier guion bajo.
2. Convertir todas las edades en números enteros.
3. Separar todos los nombres y apellidos en una sublista.

Hagámoslo ahora una función para que podamos usarla para fijar a cualquier cliente. Nombra a tu función `clean_user`. Debe recibir una lista con toda la información del cliente (user_info), así como dos enteros. Uno de ellos señala el índice del nombre del cliente y el otro es el índice de la edad del cliente en la lista. Debe devolver la lista limpia después de aplicar todos los cambios anteriores.

```python
def clean_user(user_info, name_index, age_index):
    """
    Limpia la información de un usuario: normaliza el nombre y convierte la edad a entero.

    Args:
        user_info: Lista con la información del usuario
        name_index: Índice donde se encuentra el nombre
        age_index: Índice donde se encuentra la edad

    Returns:
        Lista con la información del usuario limpia
    """
    # Paso 1: elimina del nombre espacios iniciales y finales, así como guiones
    user_name_1 = user_info[name_index].strip().replace('_', ' ')

    # Paso 2: convierte la edad en entero
    user_age_1 = int(user_info[age_index])

    # Paso 3: separa el nombre y el apellido en una sublista
    user_name_1 = user_name_1.split()

    # Prepara la lista con la información completa del usuario
    # Reemplaza el nombre y la edad originales con los datos limpios
    user_info[name_index] = user_name_1
    user_info[age_index] = user_age_1

    return user_info


# Prueba la función
test_user = ['32415', ' mike_reed ', 32.0, [
    'ELECTRONICS', 'SPORT', 'BOOKS'], [894, 213, 173]]
name_index = 1
age_index = 2

print("Usuario original:")
print(['32415', ' mike_reed ', 32.0, [
      'ELECTRONICS', 'SPORT', 'BOOKS'], [894, 213, 173]])
print("\nUsuario limpio:")
print(clean_user(test_user, name_index, age_index))
```

# Paso 4: Función clean_user Mejorada

Ahora, completemos el código de nuestra función `clean_user` para limpiar la categoría:
1. Añade otro parámetro con el índice de categorías.
2. Pon todos los nombres en minúsculas antes de aplicar "strip" y "replace".

Después, crea un bucle y aplica tu función a toda la lista de usuarios, agregando tus resultados a la lista `users_clean`.

```python
def clean_user(user_info, name_index, age_index, cat_index):
    """
    Limpia completamente la información de un usuario: nombre, edad y categorías.

    Args:
        user_info: Lista con la información del usuario
        name_index: Índice donde se encuentra el nombre
        age_index: Índice donde se encuentra la edad
        cat_index: Índice donde se encuentran las categorías

    Returns:
        Lista con la información del usuario completamente limpia
    """
    # Paso 1: pon todo en minúsculas y elimina del nombre espacios iniciales y finales, así como guiones
    user_name_1 = user_info[name_index].lower().strip().replace('_', ' ')

    # Paso 2: convierte la edad en entero
    user_age_1 = int(user_info[age_index])

    # Paso 3: separa el nombre y el apellido en una sublista
    user_name_1 = user_name_1.split()

    # Paso 4: poner categorías en minúsculas
    categories_low = []
    for category in user_info[cat_index]:
        categories_low.append(category.lower())

    # Prepara la lista con la información completa del usuario
    # Reemplaza el nombre, la edad y las categorías originales con los datos limpios
    user_info[name_index] = user_name_1
    user_info[age_index] = user_age_1
    user_info[cat_index] = categories_low

    return user_info


users = [
    ['32415', ' mike_reed ', 32.0, ['ELECTRONICS', 'SPORT', 'BOOKS'], [894, 213, 173]],
    ['31980', 'kate morgan', 24.0, ['CLOTHES', 'BOOKS'], [439, 390]],
    ['32156', ' john doe ', 37.0, ['ELECTRONICS', 'HOME', 'FOOD'], [459, 120, 99]],
    ['32761', 'SAMANTHA SMITH', 29.0, ['CLOTHES',
                                       'ELECTRONICS', 'BEAUTY'], [299, 679, 85]],
    ['32984', 'David White', 41.0, ['BOOKS', 'HOME', 'SPORT'], [234, 329, 243]],
    ['33001', 'emily brown', 26.0, ['BEAUTY', 'HOME', 'FOOD'], [213, 659, 79]],
    ['33767', ' Maria Garcia', 33.0, ['CLOTHES', 'FOOD', 'BEAUTY'], [499, 189, 63]],
    ['33912', 'JOSE MARTINEZ', 22.0, ['SPORT', 'ELECTRONICS', 'HOME'], [259, 549, 109]],
    ['34009', 'lisa wilson ', 35.0, ['HOME', 'BOOKS', 'CLOTHES'], [329, 189, 329]],
    ['34278', 'James Lee', 28.0, ['BEAUTY', 'CLOTHES', 'ELECTRONICS'], [189, 299, 579]],
]

users_clean = []

# Índices de los campos
name_index = 1
age_index = 2
cat_index = 3

# Aplicar la función a todos los usuarios
for user in users:
    # Crear una copia para no modificar la lista original
    user_copy = user.copy()
    users_clean.append(clean_user(user_copy, name_index, age_index, cat_index))

print("\n" + "="*70)
print("USUARIOS LIMPIOS")
print("="*70)
for user in users_clean[:5]:  # Mostrar los primeros 5
    print(
        f"ID: {user[0]}, Nombre: {user[1]}, Edad: {user[2]}, Categorías: {user[3]}")
```

# Paso 6: Programa de Fidelización con While

La empresa quiere ofrecer descuentos a sus clientes leales. Los clientes que realizan compras por un importe total mayor a $1500 se consideran leales y recibirán un descuento.

Nuestro objetivo es crear un bucle `while` que compruebe el importe total gastado y se detenga al alcanzarlo. Para simular nuevas compras, la variable `new_purchase` genera un número entre 30 y 80 en cada iteración del bucle.

```python
from random import randint

total_amount_spent = 1280
target_amount = 1500

print(f"Cantidad inicial gastada: ${total_amount_spent}")
print(f"Meta para descuento: ${target_amount}")
print(f"\nSimulando compras adicionales...\n")

purchase_count = 0

while total_amount_spent < target_amount:
    new_purchase = randint(30, 80)  # generamos un número aleatorio de 30 a 80
    total_amount_spent += new_purchase
    purchase_count += 1
    print(
        f"Compra #{purchase_count}: ${new_purchase} - Total acumulado: ${total_amount_spent}")

print(f"\n{'='*70}")
print(f"¡Felicidades! Has alcanzado el estatus de cliente leal.")
print(f"Total gastado: ${total_amount_spent}")
print(f"Compras adicionales necesarias: {purchase_count}")
print(f"{'='*70}")
```

# Paso 10: Función Avanzada get_client_by_cat

La dirección requiere de una función que proporcione información sobre los clientes, incluyendo sus nombres, edades y gasto total, filtrada por categorías específicas.

Crearemos una función llamada `get_client_by_cat` con las siguientes especificaciones:

1. **Parámetros:**
   - **users:** una lista con los datos de los usuarios.
   - **id_index:** el índice donde está almacenado el ID del cliente.
   - **name_index:** el índice donde está almacenado el nombre del cliente.
   - **age_index:** el índice donde la edad del cliente está almacenada.
   - **category_index:** el índice donde las categorías de compras del cliente están listadas.
   - **amounts_index:** el índice donde las cantidades gastadas en cada categoría están almacenadas.
   - **filter_category:** la categoría específica por la cual se filtrará.

2. **Resultado:**
   - Una lista de sublistas, donde cada sublista contiene: ID del usuario, nombre del usuario (como lista), edad y total de gastos.

```python
def get_client_by_cat(users, id_index, name_index, age_index, category_index, amounts_index, filter_category):
    """
    Filtra clientes por una categoría específica y retorna su información.

    Args:
        users: Lista de usuarios con toda su información
        id_index: Índice del ID de usuario
        name_index: Índice del nombre de usuario
        age_index: Índice de la edad
        category_index: Índice de las categorías de compra
        amounts_index: Índice de los montos gastados
        filter_category: Categoría para filtrar

    Returns:
        Lista de usuarios filtrados con [ID, nombre, edad, gasto_total]
    """
    result = []

    # Iterar sobre cada usuario
    for user in users:
        # Verificar si la categoría deseada está en las categorías del usuario
        if filter_category in user[category_index]:
            # Extraer información del usuario
            user_id = user[id_index]
            user_name = user[name_index]
            user_age = user[age_index]

            # Calcular el gasto total
            total_spent = sum(user[amounts_index])

            # Agregar a la lista de resultados
            result.append([user_id, user_name, user_age, total_spent])

    return result


# La lista de usuarios
users = [
    ['32415', ['mike', 'reed'], 32, ['electronics', 'sport', 'books'], [894, 213, 173]],
    ['31980', ['kate', 'morgan'], 24, ['clothes', 'books'], [439, 390]],
    ['32156', ['john', 'doe'], 37, ['electronics', 'home', 'food'], [459, 120, 99]],
    ['32761', ['samantha', 'smith'], 29, [
        'clothes', 'electronics', 'beauty'], [299, 679, 85]],
    ['32984', ['david', 'white'], 41, ['books', 'home', 'sport'], [234, 329, 243]],
    ['33001', ['emily', 'brown'], 26, ['beauty', 'home', 'food'], [213, 659, 79]],
    ['33767', ['maria', 'garcia'], 33, ['clothes', 'food', 'beauty'], [499, 189, 63]],
    ['33912', ['jose', 'martinez'], 22, [
        'sport', 'electronics', 'home'], [259, 549, 109]],
    ['34009', ['lisa', 'wilson'], 35, ['home', 'books', 'clothes'], [329, 189, 329]],
    ['34278', ['james', 'lee'], 28, [
        'beauty', 'clothes', 'electronics'], [189, 299, 579]],
]

# Definir índices
id_index = 0
name_index = 1
age_index = 2
category_index = 3
amounts_index = 4

# Prueba 1: Filtrar por categoría 'electronics'
print("\n" + "="*70)
print("CLIENTES QUE COMPRARON ELECTRÓNICOS")
print("="*70)
electronics_clients = get_client_by_cat(users, id_index, name_index, age_index,
                                        category_index, amounts_index, 'electronics')

for client in electronics_clients:
    full_name = ' '.join(client[1]).title()
    print(
        f"ID: {client[0]}, Nombre: {full_name}, Edad: {client[2]}, Gasto Total: ${client[3]}")

# Prueba 2: Filtrar por categoría 'clothes'
print("\n" + "="*70)
print("CLIENTES QUE COMPRARON ROPA")
print("="*70)
clothes_clients = get_client_by_cat(users, id_index, name_index, age_index,
                                    category_index, amounts_index, 'clothes')

for client in clothes_clients:
    full_name = ' '.join(client[1]).title()
    print(
        f"ID: {client[0]}, Nombre: {full_name}, Edad: {client[2]}, Gasto Total: ${client[3]}")

# Prueba 3: Filtrar por categoría 'sport'
print("\n" + "="*70)
print("CLIENTES QUE COMPRARON ARTÍCULOS DEPORTIVOS")
print("="*70)
sport_clients = get_client_by_cat(users, id_index, name_index, age_index,
                                  category_index, amounts_index, 'sport')

for client in sport_clients:
    full_name = ' '.join(client[1]).title()
    print(
        f"ID: {client[0]}, Nombre: {full_name}, Edad: {client[2]}, Gasto Total: ${client[3]}")
```
